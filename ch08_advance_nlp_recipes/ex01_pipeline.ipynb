# Pipeline
This example demonstrates how to create a NLP pipeline.

```python
import nltk
import threading
import queue
import feedparser
import uuid
```

```python
threads = []
queues = [queue.Queue(), queue.Queue()]
```

```python
def extract_words():
    url = 'https://www.straitstimes.com/news/singapore/rss.xml'
    feed = feedparser.parse(url)
    for entry in feed['entries']:
        text = entry['title']
        if 'ex' in text:
            continue
        words = nltk.word_tokenize(text)
        data = {'uuid': uuid.uuid4(), 'input': words}
        queues[0].put(data, True)
        print('>> {}: {}'.format(data['uuid'], text))
```

```python
def extract_pos():
    while True:
        if queues[0].empty():
            break
        else:
            data = queues[0].get()
            words = data['input']
            postags = nltk.pos_tag(words)
            queues[0].task_done()
            queues[1].put({'uuid': data['uuid'], 
                          'input': postags}, True)
```

```python
def extract_ne():
    while True:
        if queues[1].empty():
            break
        else:
            data = queues[1].get()
            postags = data['input']
            queues[1].task_done()
            chunks = nltk.ne_chunk(postags, binary=False)
            print('<< {}:'.format(data['uuid']), end='')
            for path in chunks:
                try:
                    label = path.label()
                    print(path, end=', ')
                except:
                    pass
            print()
```

```python
def run_program():
    e = threading.Thread(target=extract_words())
    e.start()
    threads.append(e)
    
    p = threading.Thread(target=extract_pos())
    p.start()
    threads.append(p)
    
    n = threading.Thread(target=extract_ne())
    n.start()
    threads.append(n)
    
    queues[0].join()
    queues[1].join()
    
    for t in threads:
        t.join()
```

```python
run_program()
```
