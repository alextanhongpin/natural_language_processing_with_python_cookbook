## Explore frequency distribution operations on one of the web and chat text corpus files

```python
!pip3 install matplotlib
import nltk
nltk.download('webtext')
from nltk.corpus import webtext
```

```python
fileids = webtext.fileids()
fileids
```

```python
fileid = 'singles.txt'
singles = webtext.words(fileid)
fdist = nltk.FreqDist(singles)
fdist
```

```python
print(f'Count of the maximum appearing token "{fdist.max()}": {fdist[fdist.max()]}')
```

```python
print(f'Total number of distinct tokens in the bag: {fdist.N()}')
```

```python
print(f'The top 10 most common words in the bag')
fdist.most_common(10)
```

```python
print('Frequency distribution on Personal Advertisements')
print(fdist.tabulate())
```

```python
fdist.plot(cumulative=True)
```
